```python
# QNA 더미 데이터 생성하기
# 총 27개
# Column = NO, WRITER, TITLE, CONTENT, CREATE_DATE, MODIFY_DATE, HITS, DELYN

import csv
import random
from datetime import datetime, timedelta

# 더미 데이터 생성 함수
def generate_dummy_data(num_rows):
    data = []
    writer_names = ['hanj12', 'kimd3', 'lees123', 'kimj89', 'choiy67', 'parkt12', 'kangj45', 'hwangr1', 'limj456', 'leew23',
                    'choib3', 'limy56', 'kimh123', 'parkj78', 'kimj91', 'leee123', 'kangy45', 'lims123', 'leew78', 'kimd123',
                    'kimh34', 'choiy345', 'leek56', 'parkh67', 'limj23', 'kimt23', 'lees23']
    
    for i in range(1, num_rows + 1):
        writer = random.choice(writer_names)
        title = "음식 관련 문의 - 레시피"
        content = "오늘은 무엇을 먹을까요? " + random.choice(["일식", "중식", "양식"]) + " 메뉴 추천 부탁드립니다."
        create_date = datetime.now() - timedelta(days=random.randint(0, 1000))
        create_date_str = create_date.strftime("%Y-%m-%d %H:%M:%S")
        row = [i, writer, title, content, create_date_str, "", 0, "N"]
        data.append(row)
    
    return data

# CSV 파일로 저장하는 함수
def save_to_csv(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["NO", "WRITER", "TITLE", "CONTENT", "CREATE_DATE", "MODIFY_DATE", "HITS", "DELYN"])
        writer.writerows(data)

# 더미 데이터 생성
num_rows = 27  # 생성할 데이터 행 수
dummy_data = generate_dummy_data(num_rows)

# CSV 파일로 저장
csv_filename = "QNA_DummyDatas.csv"
save_to_csv(dummy_data, csv_filename)
print(f"더미 데이터가 {csv_filename} 파일로 저장되었습니다.")
```
